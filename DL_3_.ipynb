```python
import numpy as np
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([0,0,0,1])
w = np.zeros(2)
b = 0
epochs = 10
print("training perception for AND gate\n")
for epoch in range(epochs):
  total_error = 0
  print(f"epoch {epoch + 1} ")
  for x, arget in zip(X,Y):
    z = np.dot(x,w) + b
    Y_pred = 1 if z >= 0 else 0
    error = arget - Y_pred
    total_error += abs (error)
    w = w + error*x
    b = b + error
    print(f"input;{x},target:{arget},prediction:{Y_pred}, error:{error}")
    print("weights:", w, "bias:", b)
    print("-" * 40)
    if total_error == 0:
      print("training converged!")
      break
print("upendra\n443")
```
