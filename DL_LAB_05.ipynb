```python
import numpy as np

def step(x):
    return 1 if x >= 0 else 0

def neuron(x1, x2, w1, w2, b):
    z = w1*x1 + w2*x2 + b
    return step(z)

inputs = [(0,0), (0,1), (1,0), (1,1)]

w1_or, w2_or, b_or = 1, 1, -0.5

w1_and, w2_and, b_and = 1, 1, -1.5

w1_nand, w2_nand, b_nand = -1, -1, 1.5

print("OR Gate:")
for x1, x2 in inputs:
    print(f"{x1}, {x2} -> {neuron(x1, x2, w1_or, w2_or, b_or)}")

print("\nAND Gate:")
for x1, x2 in inputs:
    print(f"{x1}, {x2} -> {neuron(x1, x2, w1_and, w2_and, b_and)}")

print("\nNAND Gate:")
for x1, x2 in inputs:
    print(f"{x1}, {x2} -> {neuron(x1, x2, w1_nand, w2_nand, b_nand)}")
```
